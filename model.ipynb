```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import pickle as pkl
import difflib
```

```python
vectorizer=CountVectorizer()
```

```python
vectorizer_file= open('vectorizer.pkl', 'wb')
pkl.dump(vectorizer, vectorizer_file)                     
vectorizer_file.close()
```

```python
final_df=pd.read_csv('final_df.csv')
```

```python
final_df.fillna(' ',inplace=True)
```

```python
final_df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
vectorized_data=vectorizer.fit_transform(final_df['comb'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectorized_data)
```

```python
print(similarity)
```

```python
model_file= open('model.pkl', 'wb')
pkl.dump(similarity, model_file)                     
model_file.close()
```
